```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(energy_data), columns = energy_data.columns)
xters_df = normalised_df.drop(columns = ['Appliances'])
target_variable = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(xters_df, target_variable, test_size = 0.3, random_state =42)
```

```python
from sklearn.linear_model import LinearRegression as LR
linear_model = LR()
linear_model.fit(x_train,y_train)
outcomes = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error #getting the mean absolute error
the_MAE = mean_absolute_error(y_test,outcomes)
round(the_MAE,2)
```

```python
rss = np.sum(np.square(y_test - outcomes)) # TO GET THE RESIDUAL SUM OF SQUARES
round(rss, 2)
```

```python
#now to find the root mean square error

from sklearn.metrics import mean_squared_error  
rmse = np.sqrt(mean_squared_error(y_test,outcomes))
round(rmse,3)
```

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()  #to return the weight of every feature
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```
